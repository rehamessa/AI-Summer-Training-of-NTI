```python
#combine graphs
plt.style.use("seaborn-whitegrid")
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1) #first subplot
plt.plot(x,y_sin)
plt.title('Sine')
plt.subplot(2,1,2) #Second Subplot
plt.plot(x,y_cos)
plt.title('Cosine')
plt.show()
```

```python
#Custom like colors , marker and line style
plt.figure(figsize=(9,6),dpi=80)

#generate data 
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
C,S=np.cos(x) , np.sin(x)

plt.plot(x,C,color="blue",linewidth=1.0,label='Cosine',linestyle=":")
plt.plot(x,S,color="green",linewidth=1.0,label='Sine',linestyle="--")
plt.legend()
plt.xlim(-4,4)
plt.xticks(np.linspace(-4,4,9,endpoint=True))
plt.ylim(-1,1)
plt.yticks(np.linspace(-1,1,5,endpoint=True))
plt.show()
```

```python
#Scatter Plot

a=np.random.randint(0,20,15)
b=np.random.randint(0,20,15)
print(a)
print(b)

plt.scatter(a,b,c='green',s=150,marker='o')# plot scatter

plt.show()
```

```python
level=['Excellent','good','soso']
x=range(len(level))# x-axis
y=[1,3,2]# y-axis
plt.figure(dpi=100)# create figure

plt.bar(x,y,width=0.5,color=['g','r','b'])# plot figure
plt.xticks(x,level)

# add grid
plt.grid(linestyle="--",alpha=1)
plt.show()
```
